```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
correlation(data5,0.75) # with a threshold of 0.9 the highly corelated feature 
```

# MODEL selection 

```python
data6 = data5.copy()
data6.columns
```

```python
x_train = data6.drop(columns=['systemic_crisis'],axis=1)
y_train = data6.systemic_crisis
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x_train,y_train,test_size=0.25)
```

```python
xtrain.columns
```

```python
selected_columns = ['exch_usd','inflation_annual_cpi']
xtrain_scaled = xtrain
xtest_scaled = xtest
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain_scaled[selected_columns] =  scaler.fit_transform(xtrain[selected_columns])
xtest_scaled[selected_columns] = scaler.fit_transform(xtest[selected_columns])
```

```python
xtrain_scaled.sample(4)
```

```python
# Model select
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
models = {
    'LRC':{
        'model':LogisticRegressionCV(),
        'params':{}
    },
    'RFC':{
        'model':RandomForestClassifier(),
        'params':{
            "n_estimators":[5,10,20,30,40,50,60,100]
        }
    },
    'SVC':{
        'model':SVC(),
        'params':{
            'C':[1,5,10,20],
            'kernel':['linear','rbf'],
            'gamma':['auto']
            
        }
    }
}
```

```python
scores = []
for model_name, model_info in models.items():
    cv = GridSearchCV(model_info['model'], model_info['params'], cv=5, return_train_score=False)
    cv.fit(xtrain_scaled, ytrain)  # Assuming xtrain and ytrain are defined
    scores.append({
        'model': model_name,
        'best_score': cv.best_score_,
        'best_params': cv.best_params_
    })
```

```python
scores # from this list you realise that the best model is the logistic regression with a score of 91 %
```
